```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Country and population
plt.bar(df['country'], df['population'])
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Country and Population')
```

```python
plt.bar(df['continent'], df['population'])
plt.xlabel('Continent')
plt.ylabel('Population')
plt.title('Continent and Population')
```

```python
c_2000=df.loc[df['year']==1987]
c_2000
```

```python
c_1997=df.loc[df['year']==1997]
c_1997
```

```python
Merged=pd.merge(c_2000, c_1997,on='country')
Merged
```

```python
##Drop year x and y
Merged.drop(['year_x', 'year_y', 'gdp_cap_x', 'gdp_cap_y', 'continent_x', 'continent_y'] , axis=1, inplace=True)
Merged
```

```python
#Difference in population between the 2 years
Merged['Pop_Diff']=Merged['population_y']-Merged['population_x']
Merged
```

```python
#Sort data according to the population
Merged=Merged.sort_values('Pop_Diff',ascending=False).head(30)
Merged
```

```python
#Display data of the coutnries
names=Merged['country']
plt.bar(names,Merged['Pop_Diff'])
plt.xlabel('Country')
plt.ylabel('Population Difference')
plt.title('Population Difference of the Countries')
plt.xticks(rotation=-80)
plt.show()
```
